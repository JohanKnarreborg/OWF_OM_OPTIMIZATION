```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_f = np.load('2021_2023_fake_forecast_from12oclock.npy')
weather_f = weather_f[0:]
```

```python
weather_f.shape
```

```python
weather_df = pd.read_csv('2019_2023_final.csv',index_col=0)
weather_df = weather_df.iloc[weather_df.index >= '2021-01-01']
weather_df.head()
```

```python

for day in range(0,10):
    x = np.arange(0,24)
    plt.plot(x,weather_f[day,12+0:12+24],label='forecast')
    plt.plot(x,weather_df['wind_speed'][day*24:day*24+24],label='real')
    plt.legend()
    plt.show()
```
